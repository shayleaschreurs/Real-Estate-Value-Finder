```python
# Importing libraries and modules

import os
import pandas as pd
import numpy as np
import hvplot.pandas
import json
import requests
from dotenv import load_dotenv

%matplotlib inline
```

```python
load_dotenv()
```

```python
realty_mole_api_key = os.getenv("Realty_Mole_API")
```

```python
# Initial API call to Realty Mole Property - "Property Sale Listings"

url = "https://realty-mole-property-api.p.rapidapi.com/saleListings"

querystring = {"city":"Austin","state":"TX","limit":"200"}

headers = {
	"X-RapidAPI-Key": realty_mole_api_key,
	"X-RapidAPI-Host": "realty-mole-property-api.p.rapidapi.com"
}

API_response1 = requests.request("GET", url, headers=headers, params=querystring)

print(API_response1.text)
```

```python
# Viewing API data in JSON format for easier viewing

salelistings_as_json = API_response1.json()
salelistings_as_json
```

```python
# Creating a new JSON object with API dataset

salelistings_as_json_object = pd.json_normalize(salelistings_as_json)
salelistings_as_json_object
```

```python
# Re-arranging the dataframe with relevant information

salelistings_DF = salelistings_as_json_object[['formattedAddress','propertyType','price','squareFootage','bedrooms','bathrooms', 'daysOnMarket', 'latitude', 'longitude']]
salelistings_DF

salelistings_DF_with_pricePerSqFt = salelistings_DF.assign(pricePerSqFt = salelistings_DF['price'] / salelistings_DF['squareFootage'])
salelistings_DF_with_pricePerSqFt

salelistings_DF = salelistings_DF_with_pricePerSqFt[['formattedAddress','propertyType','price','squareFootage','pricePerSqFt','bedrooms','bathrooms', 'daysOnMarket', 'latitude', 'longitude']]
salelistings_DF

salelistings_DF['pricePerSqFt'] = salelistings_DF['pricePerSqFt'].round(2)

salelistings_DF
```

```python
# Selecting only Single Family listings to analyze

selected_listings = salelistings_DF.loc[salelistings_DF["propertyType"] == "Single Family"]
selected_listings
```

```python
# Export dataframes to CSVs that can be used in other Jupyter Notebooks

salelistings_DF.to_csv("Project 1 - Raw Sales Listings v4.csv", sep='\t')
selected_listings.to_csv("Project 1 - Selected Listings v4.csv", sep='\t')
```
